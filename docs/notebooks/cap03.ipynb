# Três Distribuições Fundamentais: Binomial, Gaussiana e Poisson

```python
import sys
sys.path.append("../..")

import numpy as np
import utils.justdoit as jdi

import matplotlib.pyplot as plt
plt.rcParams.update({
    "xtick.top": True,
    "ytick.right": True,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "font.size": 12,
    "font.family": "Lato",
    "axes.labelsize": 12,
    "axes.titlesize": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "xtick.minor.visible": True,
    "ytick.minor.visible": True
    })
```

## A Distribuição Binomial

### Derivação da Distribuição Binomial

```python
N: int = 10                 # número total de tentativas
p1: float = 0.5             # probabilidades de sucesso (p1 e p2)
p2: float = 0.1
 
# valores de n de 0 a N (número de sucessos)
n: np.ndarray = np.arange(0, N + 1, 1)

# Funções de massa de probabilidade (fmp1 e fmp2)
fmp1: np.ndarray = jdi.binomial_fmp(n, N, p1)   
fmp2: np.ndarray = jdi.binomial_fmp(n, N, p2)

_, ax = plt.subplots()
ax.plot(n, fmp1, "-o", label="$N = 10$, $p = 0.5$")
ax.plot(n, fmp2, "-o", label="$N = 10$, $p = 0.5$")

ax.set(xlabel="$n$", ylabel="Probabilidade $P_N(n)$")
ax.axis([0, 10, -0.01, 0.4])
ax.set_yticks(np.arange(0, 0.5, 0.1))

plt.legend()
plt.savefig("../../tex/Figuras/3-1.pdf", bbox_inches="tight")
plt.show()
```

## A Distribuição Guassiana

```python
N: int = 10           # Número de amostras
mu1: int = 5          # Médias (mu1 e mu2)
mu2: int = 0
sigma1: float = 1.58  # Desvios padrão (sigma1 e sigma2)
sigma2: float = 1.0
 
# Conjunto de valores para os quais a FDP será calculada.                
x: np.ndarray = np.linspace(-5, 10, 100)     

# FDP guassiana
normal1: np.ndarray = jdi.normal_fdp(x, mu=mu1, sigma=sigma1)    
normal2: np.ndarray = jdi.normal_fdp(x, mu=mu2, sigma=sigma2) 

_, ax = plt.subplots()
ax.plot(x, normal1, label="$\mu = 5, \sigma^2 = 1.58$")
ax.plot(x, normal2, label="$\mu = 0, \sigma^2 = 1$")

ax.set(xlabel="$x$", ylabel="Função de distribuição $f(x)$")
ax.axis([-5, 10, -0.01, 0.41])
ax.set_yticks(np.arange(0, 0.5, 0.1))

plt.legend()
plt.savefig("../../tex/Figuras/3-2.pdf", bbox_inches="tight")
plt.show()
```

## A Distribuição de Poisson

### Momentos e Propriedades da Distribuição de Poisson

```python
# Conjunto de valores para os quais a FDP será calculada.                
n: np.ndarray = np.arange(0, 11)     
mu1: int = 5          # Médias (mu1 e mu2)
mu2: int = 1   

# FMP de Poisson
poisson1: np.ndarray = jdi.poisson_fmp(n, mu=mu1)    
poisson2: np.ndarray = jdi.poisson_fmp(n, mu=mu2) 

_, ax = plt.subplots()
ax.plot(n, poisson1, "-o", label=f"$\mu$ = {mu1}")
ax.plot(n, poisson2, "-o", label=f"$\mu$ = {mu2}")

ax.set(xlabel="$n$", ylabel="Probabilidade $P(n)$")
ax.axis([0, 10, -0.01, 0.4])
plt.legend()
plt.savefig("../../tex/Figuras/3-3.pdf", bbox_inches="tight")
plt.show()
```

## Comparação das Distribuições Binomial, Gaussiana e de Poisson

```python
n: np.ndarray = np.arange(0, 12)         
x: np.ndarray = np.linspace(-2, 12, 100)    

# FMP (BINOMIAL)
fmp1: np.ndarray = jdi.binomial_fmp(n, N=10, p=0.5)   
fmp2: np.ndarray = jdi.binomial_fmp(n, N=10, p=0.1)

# FDP (GUASSIANA)
normal1: np.ndarray = jdi.normal_fdp(x, mu=5, sigma=2.5)    
normal2: np.ndarray = jdi.normal_fdp(x, mu=1, sigma=1.0)

# FMP (POISSON)
poisson1: np.ndarray = jdi.poisson_fmp(n, mu=5)    
poisson2: np.ndarray = jdi.poisson_fmp(n, mu=1)

_, ax = plt.subplots()
ax.plot(n, fmp1, "-o", label="Binomial ($n$=10, $p$=0.5)", c="tab:blue")
ax.plot(n, fmp2, "--o", label="Binomial ($n$=10, $p$=0.1)", c="tab:blue", mfc='none')
ax.plot(x, normal1, label="Normal ($\mu$=5, $\sigma$=2.5)", c="tab:orange")
ax.plot(x, normal2, "--", label="Normal ($\mu$=1, $\sigma$=1)", c="tab:orange")
ax.plot(n, poisson1, "-o", label="Poisson ($\mu$=5)", c="tab:green")
ax.plot(n, poisson2, "--o", label="Poisson ($\mu$=1)", c="tab:green", mfc='none')

ax.set(xlabel="$n$ ou $x$", ylabel="Função de probabilidade")
ax.axis([-2, 12, -0.01, 0.4])
plt.legend()
plt.savefig("../../tex/Figuras/3-4.pdf", bbox_inches="tight")
plt.show()
```
