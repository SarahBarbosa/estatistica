## A Distribuição Binomial

### Derivação da Distribuição Binomial

```python
N: int = 10                 # número total de tentativas
p1: float = 0.5             # probabilidades de sucesso (p1 e p2)
p2: float = 0.1
 
n_valores: np.ndarray = np.arange(0, N + 1, 1)           # valores de n de 0 a N (número de sucessos)
fmp1: np.ndarray = dist.binomial_fmp(n_valores, N, p1)   # Funções de massa de probabilidade (fmp1 e fmp2)
fmp2: np.ndarray = dist.binomial_fmp(n_valores, N, p2)

_, ax = plt.subplots()
ax.plot(n_valores, fmp1, "-o", label="$N = 10$, $p = 0.5$")
ax.plot(n_valores, fmp2, "-o", label="$N = 10$, $p = 0.5$")

ax.set(xlabel="$n$", ylabel="Probabilidade $P_N(n)$")
ax.axis([0, 10, -0.01, 0.4])
ax.set_yticks(np.arange(0, 0.5, 0.1))

plt.legend()
plt.savefig("../../tex/Figuras/3-1.pdf", bbox_inches="tight")
plt.show()
```

## A Distribuição Guassiana

```python
N: int = 10           # Número de amostras
mu1: int = 5          # Médias (mu1 e mu2)
mu2: int = 0
sigma1: float = 1.58  # Desvios padrão (sigma1 e sigma2)
sigma2: float = 1.0
 
x: np.ndarray = np.linspace(-5, 10, 100)    # Conjunto de valores para os quais a FDP será calculada.                           
normal1: np.ndarray = dist.normal_fdp(x, mu=mu1, sigma=sigma1)    # FDP guassiana
normal2: np.ndarray = dist.normal_fdp(x, mu=mu2, sigma=sigma2)    # FDP guassiana      

_, ax = plt.subplots()
ax.plot(x, normal1, label="$\mu = 5, \sigma^2 = 1.58$")
ax.plot(x, normal2, label="$\mu = 0, \sigma^2 = 1$")

ax.set(xlabel="$x$", ylabel="Função de distribuição $f(x)$")
ax.axis([-5, 10, -0.01, 0.41])
ax.set_yticks(np.arange(0, 0.5, 0.1))

plt.legend()
plt.savefig("../../tex/Figuras/3-2.pdf", bbox_inches="tight")
plt.show()
```
